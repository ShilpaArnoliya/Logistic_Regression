```python
# converting categorical data inti numeric
#1st we convert yes no into 0&1 i.e default, housing & y column
B['default']=B['default'].apply(lambda x:0 if x=='no' else 1)
B['housing']=B['housing'].apply(lambda x:0 if x=='no' else 1)
B['loan']=B['loan'].apply(lambda x:0 if x=='no' else 1)
B['y']=B['y'].apply(lambda x:0 if x=='no' else 1)
```

```python
# now we convert multipul categorical cvalues columns i.e job marital education contact poutcome
def month_converter(mnth):
    months = ['jan', 'feb','mar','apr','may','jun','jul','aug','sep','oct','nov','dec']
    return months.index(mnth)+1
    
B['month']=B['month'].apply(lambda x:month_converter(x))
```

```python
B.head(10)
```

```python
def job_converter(j):
    job_column = ['management', 'technician', 'entrepreneur', 'blue-collar',
       'unknown', 'retired', 'admin.', 'services', 'self-employed',
       'unemployed', 'housemaid', 'student']
    return job_column.index(j)
    
B['job']=B['job'].apply(lambda x:job_converter(x))
```

```python
B.head(10)
```

```python
B.dtypes
```

```python
def mrg_converter(m):
    mritl_column = ['divorced', 'single', 'married']
    return mritl_column.index(m)
    
B['marital']=B['marital'].apply(lambda x:mrg_converter(x))
```

```python
B.head(10)
```

```python
def edu_converter(E):
    edu_column = ['unknown','primary','secondary','tertiary']
    return edu_column.index(E)
    
B['education']=B['education'].apply(lambda x:edu_converter(x))
```

```python
B.head(10)
```

```python
def con_converter(c):
    cont_column = ['unknown', 'cellular', 'telephone']
    return cont_column.index(c)
    
B['contact']=B['contact'].apply(lambda x:con_converter(x))
```

```python
B.head(10)
```

```python
def pout_converter(c):
    pout_column = ['failure','success','other','unknown']
    return pout_column.index(c)
    
B['poutcome']=B['poutcome'].apply(lambda x:pout_converter(x))
```

```python
# Model building 
from sklearn.linear_model import LogisticRegression

B.shape
X = B.iloc[:,:16]
Y = B.iloc[:,16:17]
classifier = LogisticRegression()
classifier.fit(X,Y)

#classifier.coef_ # coefficients of features (B0,B1)
#classifier.predict_proba (X) # Probability values 
```

```python
y_pred = classifier.predict(X)
B.insert(loc=0, column='Pred', value=y_pred)
```

```python
B.Pred.head(10)
```

```python
y_prob = pd.DataFrame(classifier.predict_proba(X.iloc[:,:]))
new_df = pd.concat([B,y_prob],axis=1)
B.head(10)
```
